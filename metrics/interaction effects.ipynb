```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')

nobs = 100
x1 = np.random.normal(0,2,nobs)
x2 = np.random.normal(0,2,nobs)
d = np.random.choice([0,1],nobs)
e = np.random.normal(0,1,nobs)
y = d*x1 + (1-d)*x2 + e

ax.scatter(x2, y, x2)
plt.show()
plt.scatter(x1,y)
plt.scatter(x2,y)
plt.show()
```
