```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.sandbox.regression.gmm import IV2SLS

# Set the random seed for reproducibility
np.random.seed(0)

# Number of samples
n = 1000

# Function to simulate the data
def generate_data(n):
    u = np.random.normal(0, 1, n)
    e1 = np.random.normal(0, 1, n)
    e2 = np.random.normal(0, 1, n)
    e3 = np.random.normal(0, 1, n)
    
    z1 = 3 + 5 * u + e1
    z2 = 1 + 6 * u + e2
    
    x = 2* z1 + z2 + 3*e3
    y = 2 +  3 * x + 10 * u
    
    return y, x, z1, z2, u

# Generate the data
y, x, z1, z2, u = generate_data(n)

def ols_estimation(y, x):
    x_with_const = sm.add_constant(x)
    model = sm.OLS(y, x_with_const)
    results = model.fit()
    print("OLS Estimation:")
    print(results.summary())
    return results

def two_sls_estimation(y, x, instruments):
    x_with_const = sm.add_constant(x)
    instruments_with_const = sm.add_constant(instruments)
    model = IV2SLS(y, x_with_const, instruments_with_const)
    results = model.fit()
    print("2SLS Estimation:")
    print(results.summary())
    return results

# Perform OLS estimation
ols_results = ols_estimation(y, x)

# Perform 2SLS estimation with z1 as an instrument
print("\n2SLS using z1 as an instrument")
two_sls_results_z1 = two_sls_estimation(y, x, z1)

# Perform 2SLS estimation with z2 as an instrument
print("\n2SLS using z2 as an instrument")
two_sls_results_z2 = two_sls_estimation(y, x, z2)

# Perform 2SLS estimation with both z1 and z2 as instruments
print("\n2SLS using both z1 and z2 as instruments")
two_sls_results_z1_z2 = two_sls_estimation(y, x, np.column_stack((z1, z2)))

#TODO probably possible to estimate e1 and e2... should do that and then regress those on x in 2 stage 1...
```
