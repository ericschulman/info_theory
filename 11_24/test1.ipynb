```julia
using Optim
using PyPlot
```

```julia
for ALPHA =[-1.45] 
    C = .2
    obj(x,y) = (x-C)*exp(ALPHA*x)/( 1+ exp(ALPHA*y) + exp(ALPHA*x) )
    results = []
    ys =[]
    for y=0:0.2:1

        obj2(x) = -1*obj(x[1],y)
        res = optimize(obj2,[0.0],BFGS())
        res = res.minimizer
        results = append!(results,res[1])
        ys = append!(ys,y)
    end
    plot(ys,results,label="alpha=$(ALPHA)")
end
print(results[1])

plot(ys,ones(6))
xlabel("Rival price (expected)")
ylabel("Price")
legend()
show()

```

```julia
for C =[.1,.15,.2] 
    ALPHA = -3
    obj(x,y) = (x-C)*exp(ALPHA*x)/( 1+ exp(ALPHA*y) + exp(ALPHA*x) )
    results = []
    ys =[]
    for y=0:0.2:1

        obj2(x) = -1*obj(x[1],y)
        res = optimize(obj2,[0.0],BFGS())
        res = res.minimizer
        results = append!(results,res[1])
        ys = append!(ys,y)
    end
    plot(ys,results,label="cost=$(C)")
end

xlabel("Rival price (expected)")
ylabel("Price")
legend()
show()
```
