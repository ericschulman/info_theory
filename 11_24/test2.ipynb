```julia
using Optim
using PyPlot
```

```julia
ALPHA =-3

C = .2
obj(x,y) = (x-C)*exp(ALPHA*x)/( 1+ exp(ALPHA*y) + exp(ALPHA*x) )
results = []
ys =[]
for y=0:0.2:1

    obj2(x) = -1*obj(x[1],y)
    res = optimize(obj2,[0.0],BFGS())
    res = res.minimizer
    results = append!(results,res[1])
    ys = append!(ys,y)
end
plot(ys,results,label="p1=$(ALPHA)")
plot(results,ys,label="p2=$(ALPHA)")
plot(ys,ys,label="p2=$(ALPHA)")
print(results[1])

#plot(ys,ones(6))
xlabel("Rival price (expected)")
ylabel("Price")
legend()
show()

```
