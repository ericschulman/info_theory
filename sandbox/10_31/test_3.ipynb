```julia
import Pkg
using Optim
using PyPlot
using LinearAlgebra
using FFTW
```

```julia
T = 10
signal = collect(1:T)
println(fft(signal))
```

```julia
T = 10
signal = collect(1:T)
freq = fft(signal) 
println(freq)
println(freq|> fftshift )
filter = freq[:]

for i=1:T
    if mod(i,2) !=0
        filter[i] =0
    end
end

filter = ifft(filter)
print(abs.(filter))

plot(signal,signal,label="original signal")
plot(signal,freq,label="fft")
plot(signal,filter,label="filter")
legend()
show()
```
