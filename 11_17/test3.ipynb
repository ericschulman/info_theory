```julia
using Pkg
using Optim
using PyPlot
using LinearAlgebra
using FFTW
using Random
```

```julia
T = 1000
xs = zeros(T)
as = zeros(T)

x = 0
a = 0
es = zeros(T)

f(x,e) = mod(x+e,5.0)

for t=2:T
    
    e = randsubseq([1],0.8)
    e = Integer(e != [])
    
    x = f(x,e)
    a = Integer( (x == 0) & (xs[t-1]!= 0) ) 
    
    xs[t] = x
    es[t] = e
    as[t] =a
    
end

figure(1)
plot(collect(1:T),as)
```

```julia
fs = fft(as.- sum(as)/T)
fs = reinterpret(Float64, fs)
fre = @view fs[1:2:end-1]
fim = @view fs[2:2:end]

#plot(collect(1:T),fre,label="re")
plot(collect(1:T),fim,label="im")
```
