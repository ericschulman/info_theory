```julia
function calc_pr(params, n)
    prs = zeros(3)
    for x = 1:3
        u1 = params[2]*n[x,1] + params[3]*(1-n[x,1])
        u2 = params[2]*n[x,2] + params[3]*(1-n[x,2])
        prs[x] = params[1]*exp(u1)/(1+exp(u1)) + (1-params[1])*exp(u2)/(1+exp(u2))
    end
    return prs
end

n11 = .5
n12 = .6
n13 = .7

n21 = .3
n22 = .4
n23 = .5

n = [ n11 n21 ; n12 n22 ;  n13 n23 ]

print(calc_pr([.5 3 -1], n))
```
