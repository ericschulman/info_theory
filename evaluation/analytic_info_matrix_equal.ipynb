```python
from sympy import *
```

```python
mu = Symbol('mu')
sigma = Symbol('sigma')
x = Symbol('x')
n = Symbol('n')


#log_likelihood = n/2*ln(2*pi) - n/2*ln(sigma**2) - 1/(2*sigma**2) * (x-mu)**2


likelihood = 1/(sqrt(2*pi)*sigma)*exp(-1/2*((x-mu)/sigma)**2)
log_likelihood = ln(likelihood)
```

```python
theta = [mu,sigma]
grad = Matrix(1,2, [log_likelihood.diff(x) for x in theta])
hess = Matrix(2,2,[log_likelihood.diff(x).diff(y) for x in theta for y in theta])

expectation_1 = integrate(grad.transpose()*(grad)*likelihood,x)

print(expectation_1)
```

```python
expectation_2 = integrate(hess*likelihood,x)
print(expectation_2)
```
