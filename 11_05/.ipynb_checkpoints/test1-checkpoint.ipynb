```python
import pandas as pd
import math
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
data = pd.read_csv('bus1234.csv')
data['day'] = 1
data['day'] = pd.to_datetime(data[['year','month','day']])
```

```python
for bid in pd.unique(data['id']):
    if bid:
        new_data = data[data['id'] == bid]
        plt.figure(1, figsize=(16, 6))
        plt.plot(new_data['day'], new_data['replace'],label= 'id %s'%bid)
        plt.figure(2, figsize=(16, 6))
        plt.plot(new_data['day'], new_data['miles']/1e6,label= 'id %s'%bid)


#plt.legend()
plt.figure(1)
#plt.legend()
plt.show()
```

```python
model = sm.OLS(data['replace'], sm.add_constant(data['miles']/1e6))
result = model.fit()
result.summary()

model = sm.Logit(data['replace'], sm.add_constant(data['miles']/1e6))
result = model.fit()
result.summary()
```

```python
theta1 = -.58
RC = -10.47

util = data['miles']*theta1*(1-data['replace'])/1e6 + data['replace']*RC
print( util.mean() )

print((data['miles']*(1-data['replace'])/1e6*theta1 ).mean())
print(data['replace'].mean()*RC )
```
